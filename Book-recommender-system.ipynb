## Popularity Based Recommender System

```python
ratings_with_name = ratings.merge(books, on="ISBN")
```

```python
ratings_with_name
```

```python
ratings_with_name.dtypes
```

```python
num_ratings_df = ratings_with_name.groupby("Book-Title").count()["Book-Rating"].reset_index()
num_ratings_df.rename(columns={"Book-Rating":"num_ratings"}, inplace=True)
num_ratings_df
```

```python
avg_ratings_df = ratings_with_name[["Book-Rating","Book-Title"]].groupby("Book-Title").mean().reset_index()
avg_ratings_df.rename(columns={"Book-Rating":"avg_rating"}, inplace=True)
avg_ratings_df
```

```python
popular_df = num_ratings_df.merge(avg_ratings_df, on="Book-Title")
popular_df
```

```python
popular_df = popular_df[popular_df["num_ratings"] >= 250].sort_values("avg_rating", ascending=False).iloc[0:50]
popular_df
```

```python
popular_df = popular_df.merge(books, on="Book-Title").drop_duplicates("Book-Title")[["Book-Title","Book-Author","Image-URL-L","num_ratings","avg_rating"]]
```

## Collaborative Filtering Based Recommender System

```python
X = ratings_with_name.groupby("User-ID").count()["Book-Rating"] > 200
literate_users = X[X].index
```

```python
literate_users
```

```python
filtered_rating = ratings_with_name[ratings_with_name["User-ID"].isin(literate_users)]
```

```python
filtered_rating
```

```python
Y = filtered_rating.groupby("Book-Title").count()["Book-Rating"] >= 50
famous_books = Y[Y].index
```

```python
famous_books
```

```python
final_ratings = filtered_rating[filtered_rating["Book-Title"].isin(famous_books)]
```

```python
final_ratings
```

```python
final_ratings.drop_duplicates(inplace=True)
```

```python
pt = final_ratings.pivot_table(index="Book-Title",columns="User-ID",values="Book-Rating")
```

```python
final_ratings["Book-Title"].unique().shape
```

```python
final_ratings["User-ID"].unique().shape
```

```python
pt
```

```python
pt.fillna(0, inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score
```

```python
similarity_score.shape
```

```python
first_book = list(sorted(enumerate(similarity_score[0]), reverse=True, key=lambda x:x[1])[1:6])
first_book
```

```python
def recommend(book_name):
    index = np.where(pt.index==book_name)[0][0]
    distances = similarity_score[index]
    similar_items = list(sorted(enumerate(distances), reverse=True, key=lambda x:x[1])[1:6])
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books["Book-Title"] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Book-Title"].values))
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Book-Author"].values))
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Image-URL-L"].values))
        data.append(item)
    return data
```

```python
recommend("The Da Vinci Code")
```

```python
recommend("The Hobbit : The Enchanting Prelude to The Lord of the Rings")
```

```python
recommend("The Two Towers (The Lord of the Rings, Part 2)")
```

```python
recommend("Harry Potter and the Chamber of Secrets (Book 2)")
```
